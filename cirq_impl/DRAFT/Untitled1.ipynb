```python
import cirq
import qsimcirq

q0, q1, q2 = [cirq.LineQubit(i) for i in range(3)]
qc = cirq.Circuit(cirq.X(q0))
print(qc)

print()
print()
qc.append(cirq.measure([q2, q1, q0]))
print(qc)
```

```python
def bitstring(bits):
    return ''.join(str(b) for b in bits)
```

```python
qsimcirq.QSimSimulator().run(qc, repetitions=1000).histogram(key=[q2, q1, q0], fold_func=bitstring)
```

```python
help(qsimcirq.QSimSimulator().run(qc, repetitions=1000).histogram)
```

```python
print(getsource(qsimcirq.QSimSimulator().run(qc, repetitions=1000).histogram))
```
