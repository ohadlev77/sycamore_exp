```python
import pandas as pd

df = pd.read_csv("exp_D08.05.23_T12.51.59/circuit_n12_m14_s0_e0_pEFGH.csv", index_col=0)

weber_probs = df["qvm_weber_prob"]
ideal_probs = df["ideal_prob"]
depo_probs = df["depolarize_single_0.0016_prob"]

df
```

```python
import numpy as np

print((np.dot(ideal_probs, weber_probs) * 2**12) - 1)
print((np.dot(ideal_probs, depo_probs) * 2**12) - 1)
```
