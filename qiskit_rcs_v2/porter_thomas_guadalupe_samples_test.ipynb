```python
import pandas as pd

df = pd.read_csv("exp_data_12q_fake_gudalupe_v2/circuit_0_data.csv")
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

num_qubits = 12
hilbert_dim = 2 ** num_qubits

scaled_ideal_probs = hilbert_dim * df["ideal_prob"]

plt.hist(scaled_ideal_probs, bins=80, density=True, rwidth=0.8)

x_exp = np.linspace(0, max(scaled_ideal_probs))
plt.plot(x_exp, np.e**-(x_exp))

plt.show()
```

```python
scaled_noisy_probs = hilbert_dim * df["noisy_prob"]

plt.hist(scaled_noisy_probs, bins=80, density=True, rwidth=0.8)

x_exp = np.linspace(0, max(scaled_noisy_probs))
plt.plot(x_exp, np.e**-(x_exp))

plt.show()
```
