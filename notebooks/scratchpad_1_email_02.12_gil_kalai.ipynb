```python
import json

with open(backend_props_json, "r") as f:
    props = json.load(f)
    
props
```

```python
from qiskit import qpy

with open("../qiskit_rcs_v2/exp_data__12q__fake_guadalupe__v2/transpiled_circuits.qpy", "rb") as f:
    transpiled_circuits = qpy.load(f)
    
transpiled_circuits[1].draw("mpl", fold=-1)
```

**Not used qubits:** 0, 9, 10, 15

```python
not_used_qubits = {0, 9, 10, 15}
num_total_qubits = 16
num_used_qubits = 12

total_error = 0
for i in range(16):
    if i not in (not_used_qubits):
        current_error = props["qubits"][i][4]["value"]
        total_error += current_error
        print(f"Readout error for qubit {i} = {current_error}")
        
print()        
print(f"Average readout error for all used qubits = {total_error / num_used_qubits}")
```
